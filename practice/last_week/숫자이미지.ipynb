```python
from keras.datasets import mnist
```

```python
(X_train_0, Y_train_0), (X_test_0, Y_test_0) = mnist.load_data()
```

```python
print("학습 데이터 수 : %d"%(X_train_0.shape[0]))
print("평가 데이터 수 : %d"%(X_test_0.shape[0]))
```

```python
from tensorflow.python.keras.utils import np_utils
```

```python
X_train = X_train_0.reshape(X_train_0.shape[0], 28, 28, 1).astype('float32') / 255
X_test = X_test_0.reshape(X_test_0.shape[0], 28, 28, 1).astype('float32') / 255
```

```python
Y_train = np_utils.to_categorical(Y_train_0, 10)
Y_test = np_utils.to_categorical(Y_test_0, 10)
print(Y_train[0])
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.callbacks import ModelCheckpoint,EarlyStopping
import os
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), input_shape=(28, 28, 1), activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128,  activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
MODEL_DIR = '/Users/parkjuyong/Desktop/4-1/Data Crawling/practice/last_week/data'
if not os.path.exists(MODEL_DIR): 
    os.mkdir(MODEL_DIR)

modelpath="/Users/parkjuyong/Desktop/4-1/Data Crawling/practice/last_week/data/{epoch:02d}-{val_loss:.4f}.hdf5"
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=5)
```

```python
history = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), \
                    epochs=30, batch_size=200, verbose=0, \
                    callbacks=[early_stopping_callback,checkpointer])
```

```python
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))
```

```python
import numpy as np
```

```python
y_vloss = history.history['val_loss']

y_loss = history.history['loss']
```

```python
x_len = np.arange(len(y_loss))
plt.plot(x_len, y_vloss, marker='.', c="red", label='Testset_loss')
plt.plot(x_len, y_loss, marker='.', c="blue", label='Trainset_loss')
plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
