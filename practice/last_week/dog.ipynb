```python
import os, glob
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.python.keras.utils import np_utils
```

```python
root_dir = "/Users/parkjuyong/Desktop/4-1/Data Crawling/practice/last_week/data"
```

```python
categories = ["말티즈", "요크셔", "푸들"]
num_class = len(categories)
```

```python
image_width = 64
image_height = 64
```

```python
X = [] 
Y = [] 

for idx, category in enumerate(categories):
    image_dir = root_dir + category
    files = glob.glob(image_dir + "/*.*")        

    for i, f in enumerate(files):                
        img = Image.open(f)
        img = img.convert("RGB")
        img = img.resize((image_width, image_height))
        data = np.asarray(img)                
        X.append(data)
        Y.append(idx)
    print(f"{image_dir} : {i+1}") 

print(f"image shape : {data.shape}") 
print("X 데이터 총 개수 :", len(X)) 
print("Y 데이터 총 개수 :", len(Y)) 
```

```python
np_x = np.array(X)
np_y = np.array(Y)

print(np_x.shape) 
print(np_y.shape) 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(np_x, np_y, test_size=0.2, random_state=42)
```

```python
X_train = X_train.astype("float") / 255
X_test = X_test.astype("float") / 255
```

```python
Y_train = np_utils.to_categorical(Y_train, num_class)
Y_test = np_utils.to_categorical(Y_test, num_class)
print(Y_train[0])
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from keras.callbacks import ModelCheckpoint,EarlyStopping
```

```python
input_shape = (image_width, image_height, 3)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), input_shape=input_shape, activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128,  activation='relu'))
model.add(Dense(num_class, activation='softmax'))
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
MODEL_DIR = root_dir+'model/'
if not os.path.exists(MODEL_DIR): 
    os.mkdir(MODEL_DIR)

modelpath = MODEL_DIR+"{epoch:02d}-{val_loss:.4f}.hdf5"
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=5)
```

```python
model.fit(X_train, Y_train, validation_data=(X_test, Y_test), \
                    epochs=30, batch_size=20, verbose=0,\
                   callbacks=[early_stopping_callback,checkpointer])
```

```python
score = model.evaluate(X_test, Y_test)
print('loss =', score[0])
print('accuracy =', score[1])
```
