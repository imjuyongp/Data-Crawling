```python
# 피처 이름 파일 읽어오기
feature_name_df = pd.read_csv('/Users/parkjuyong/Desktop/4-1/Data Crawling/practice/week10/UCI_HAR_Dataset/features.txt',
                              sep = '\s+', header=None, names = ['index','feature_name'], engine='python')
```

```python
# index는 제거하고, feature_namea만 리스트로 저장
feature_name = feature_name_df.iloc[:,1].values.tolist()
```

```python
feature_name[:5]
```

```python
X_train = pd.read_csv('/Users/parkjuyong/Desktop/4-1/Data Crawling/practice/week10/UCI_HAR_Dataset/train/X_train.txt',
                      delim_whitespace = True, header = None, encoding='latin-1')
X_train.columns = feature_name
X_test = pd.read_csv('/Users/parkjuyong/Desktop/4-1/Data Crawling/practice/week10/UCI_HAR_Dataset/test/X_test.txt',
                     delim_whitespace = True, header = None, encoding='latin-1')
X_test.columns = feature_name
Y_train = pd.read_csv('/Users/parkjuyong/Desktop/4-1/Data Crawling/practice/week10/UCI_HAR_Dataset/train/y_train.txt',
                              sep = '\s+', header=None, names = ['action'], engine='python')
Y_test = pd.read_csv('/Users/parkjuyong/Desktop/4-1/Data Crawling/practice/week10/UCI_HAR_Dataset/test/y_test.txt',
                              sep = '\s+', header=None, names = ['action'], engine='python')
```

```python
X_train.head() # 작업 확인용 출력
```

```python
print(Y_train['action'].value_counts())
```

```python
label_name_df = pd.read_csv('/Users/parkjuyong/Desktop/4-1/Data Crawling/practice/week10/UCI_HAR_Dataset/activity_labels.txt',
                            sep = '\s+', header = None, names = ['index', 'label'], engine = 'python')
```

```python
# index는 제거하고, feature_name만 리스트로 저장
label_name = label_name_df.iloc[:,1].values.tolist()
```

```python
label_name
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# 결정 트리 분류 분석: 모델 생성
dt_HAR = DecisionTreeClassifier(random_state=156)
```

```python
# 결정 트리 분류 분석 : 모델 훈련
dt_HAR.fit(X_train, Y_train)
```

```python
# 결정 트리 분류 분석 : 평가 데이터에 대한 예측 수행 -> 예측결과로 Y_predict구하기
Y_predict = dt_HAR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(Y_test, Y_predict)
print('결정 트리 예측 정확도: {0:.4f}'.format(accuracy))
```

```python
print('결정 트리의 현재 하이퍼 매개변수 : \n', dt_HAR.get_params())
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'max_depth':[6,8,10,12,16,20,24]
}

grid_cv=GridSearchCV(dt_HAR, param_grid=params, scoring='accuracy', cv=5, return_train_score = True)

grid_cv.fit(X_train, Y_train)
```

```python
print('최고 평균 정확도: {0:.4f}, 최적 하이퍼 매개변수: {1}'.format(grid_cv.best_score_,grid_cv.best_params_))
```

```python
best_dt_HAR = grid_cv.best_estimator_
best_Y_predict = best_dt_HAR.predict(X_test)
best_accuracy=accuracy_score(Y_test, best_Y_predict)

print('best 결정 트리 예측 정확도: {0:.4f}'.format(best_accuracy))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
feature_importance_values = best_dt_HAR.feature_importances_
feature_importance_values_s = pd.Series(feature_importance_values, index = X_train.columns)
```

```python
feature_top10 = feature_importance_values_s.sort_values(ascending=False)[:10]
```

```python
plt.figure(figsize=(10,5))
plt.title('Feature Top 10')
sns.barplot(x=feature_top10, y = feature_top10.index)
plt.show()
```
